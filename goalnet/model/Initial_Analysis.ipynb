### Analysis

```python
def count_tasks(data_collector):
    data = data_collector.data
    active_tasks = []
    complete_tasks = []
    
    for k, d in data.iteritems():
        tasks = d["tasks"]
        active = 0
        complete = 0
        for t in tasks.values():
            if t['active']: active += 1
            else: complete += 1
        active_tasks.append(active)
        complete_tasks.append(complete)
    return (active_tasks, complete_tasks)
```

```python
active_tasks, complete_tasks = count_tasks(w.data_collector)
active_tasks = np.array(active_tasks)
complete_tasks = np.array(complete_tasks)
total_tasks = active_tasks + complete_tasks

fig, subplot = plt.subplots(figsize=(10,10))
subplot.set_title("Task Counts")
subplot.set_xlabel("Time")
subplot.set_ylabel("Count")
subplot.plot(range(400), active_tasks, range(400), complete_tasks, range(400), total_tasks)
subplot.legend(["Active_Tasks", "Complete_Tasks", "Total Tasks"], loc=(1.03, 0))
```

```python
def gini(data_collector):
    data = data_collector.data
    ginis = []
    
    for k, d in data.iteritems():
        wealths = sorted(d["wealth"].values())
        sum_iy = 0
        sum_y = 0
        n = len(wealths)
        for i, y in enumerate(wealths):
            sum_iy += i*y
            sum_y += y
        if sum_y > 0:
            G = ((2.0*sum_iy)/(n*sum_y)) - ((n + 1.0)/n)
        else:
            G = 0
        ginis.append(G)
    return ginis
```

```python
fig, subplot = plt.subplots()
subplot.set_title("Gini Coefficient")
subplot.set_xlabel("Time")
subplot.set_ylabel("Gini")
plt.plot(gini(w.data_collector))
```

```python
# Wealth v. greed at the end:
wealth = []
greed = []
for agent in w.agents.values():
    wealth.append(agent.wealth)
    greed.append(agent.greed)

fig, subplot = plt.subplots()
subplot.set_title("Agent Wealth vs Greed")
subplot.set_xlabel("Greed")
subplot.set_ylabel("Wealth")
subplot.scatter(greed, wealth)
```

```python
# Wealth v. centralization at the end:
wealth = []
centralization = []
for agent in w.agents.values():
    wealth.append(agent.wealth)
    centralization.append(agent.centralization)

    
fig, subplot = plt.subplots()
subplot.set_title("Agent Wealth vs Centralization")
subplot.set_xlabel("Centralization")
subplot.set_ylabel("Wealth")
plt.scatter(centralization, wealth)
```

```python
sizes = np.array(wealth) * 5

fig, subplot = plt.subplots()
subplot.set_title("Centralization vs Greed (size=Wealth)")
subplot.set_xlabel("Centralization")
subplot.set_ylabel("Greed")
subplot.set_xlim(0,1)
subplot.set_ylim(0,1)
plt.scatter(centralization, greed, s=sizes)
```

```python
# Propensity to help:

wealth = []
pth = []
for agent in w.agents.values():
    wealth.append(agent.wealth)
    pth.append(agent.propensity_to_help)

    
fig, subplot = plt.subplots()
subplot.set_title("Agent Wealth vs PTH")
subplot.set_xlabel("PTH")
subplot.set_ylabel("Wealth")
plt.scatter(pth, wealth)
```
